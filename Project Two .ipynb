```python
from pandas.tseries.frequencies import to_offset
```

```python
ge.loc["03-2021"].iplot(kind= "candle")
```

```python
weekly = ge.resample("W-wed").ohlc() #一週中特定挑選某一天
weekly
```

```python
weekly = ge.resample("W-Fri").agg({"Open":"first",
                                   "High":"max",
                                   "Low":"min",
                                   "Close":"last"
                                  })
weekly
```

```python
weekly.index = weekly.index - to_offset("4d")
```

```python
weekly.loc["4-2021":"5-2021"].iplot(kind= "candle")
```

# Technical Indicators


```python
qf = cf.QuantFig(df = ge.loc["2020":"2021"])
qf.add_ema(periods = 20)
qf.add_ema(periods = 100)
qf.iplot(title = "GE", name = "GE")
```

```python
qf.add_volume(colorchange=False)
qf.iplot(title = "GE", name = "GE")
```

```python
qf = cf.QuantFig(df = ge.loc["2020":"2021"])
qf.add_sma(periods = 20)
qf.iplot(title = "GE", name = "GE")
qf.add_sma(periods = 100)
qf.add_bollinger_bands(periods = 20, boll_std= 2)
qf = cf.QuantFig(df = ge.loc["5-2020":"9-2021"])
qf.iplot(title = "GE", name = "GE")
qf.add_macd()
qf.add_dmi()
```
